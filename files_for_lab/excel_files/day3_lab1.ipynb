```python
#Show a plot of the total number of responses.
display(data['Response'].unique())
data['Response'].value_counts()

data['Response'].value_counts().plot(kind='bar')
```

```python
# Show a plot of the response rate by the sales channel.
sns.set_theme(style="whitegrid")
df = sns.load_dataset("titanic")
plot_data = data
sns.countplot(data=plot_data, x="Sales Channel", hue="Response")
plt.show()
```

```python
# Show a plot of the response rate by the total claim amount.
sns.set_theme(style="whitegrid")
df = sns.load_dataset("titanic")
plot_data = data
sns.countplot(data=plot_data, x="Total Claim Amount", hue="Response")
plt.show()
```

```python
# Show a plot of the response rate by income
sns.set_theme(style="whitegrid")
df = sns.load_dataset("titanic")
plot_data = data
sns.countplot(data=plot_data, x="Income", hue="Response")
plt.show()
```

```python
#all data side by side
sns.set_theme(style="whitegrid")
df = sns.load_dataset("titanic")
plot_data = data

fig, ax =plt.subplots(1,2,3)
sns.countplot(plot_data['Sales Channel'], ax=ax[0])
sns.countplot(plot_data['Total Claim Amount'], ax=ax[1])
sns.countplot(plot_data['income'], ax=ax[2])
fig.show()
```
